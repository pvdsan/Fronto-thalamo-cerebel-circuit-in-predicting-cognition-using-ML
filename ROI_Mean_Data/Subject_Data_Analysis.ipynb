```python
def plot_gender_distribution(df1, df2, id_column='src_subject_id', gender_column='demo_sex_v2'):
    """
    Merges two DataFrames on a specified ID column and plots the gender distribution as a pie chart.
    
    Parameters:
    df1 (pd.DataFrame): First DataFrame containing subject IDs.
    df2 (pd.DataFrame): Second DataFrame containing subject IDs and gender.
    id_column (str): The column name of the subject ID in both DataFrames. Default is 'subject_id'.
    gender_column (str): The column name of the gender in df2. Default is 'gender'.
    
    Returns:
    None
    """
    # Merge the two dataframes on 'subject_id'
    merged_df = pd.merge(df1, df2, on=id_column, how='left')

    # Count the gender distribution
    gender_counts = merged_df[gender_column].value_counts()

    # Plotting the pie chart
    plt.figure(figsize=(6, 6))
    wedges, texts, autotexts = plt.pie(
        gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=90, colors=['#ff9999', '#66b3ff','yellow']
    )

    # Add legend
    plt.legend(wedges, gender_counts.index, title="Gender", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))
    
    # Add title and display the chart
    plt.title('Gender Distribution')
    plt.show()
```

```python
def plot_race_distribution(df, race):
    """
    Plots a bar chart for race distribution based on binary race columns (0/1) in the DataFrame,
    with the percentage of each race category displayed above the bars.
    
    Parameters:
    df (pd.DataFrame): DataFrame containing subject IDs and binary race columns.
    race (pd.DataFrame): DataFrame containing race information with binary columns.
    
    Returns:
    None
    """
    
    
    
    # Merging the data on the 'src_subject_id' column
    race_df = pd.merge(df, race, on='src_subject_id', how='left')
    
    # Drop the 'src_subject_id' column, as we are only interested in race columns
    race_df = race_df[['demo_race_a_p___10','demo_race_a_p___11', 'demo_race_a_p___12', 'demo_race_a_p___13', 'demo_race_a_p___14','demo_race_a_p___15','demo_race_a_p___16','demo_race_a_p___17','demo_race_a_p___18','demo_race_a_p___19','demo_race_a_p___20','demo_race_a_p___21','demo_race_a_p___22','demo_race_a_p___23','demo_race_a_p___24','demo_race_a_p___25','demo_race_a_p___77','demo_race_a_p___99']]
    
    # Summing values in each column to get counts of subjects in each race category
    race_counts = race_df.sum()

    # Dynamic extraction of column names for race labels
    race_labels = ['white', 'black', 'nat_amr','nat_als','nat_haw','guam','samoa','pacif_isl','ind','chin','filip','jap','kor','viet','othr_asian','oth_race']


    # Calculate percentages
    total_subjects = race_counts.sum()
    race_percentages = (race_counts / total_subjects) * 100

    # Plotting the bar chart
    plt.figure(figsize=(10, 6))
    bars = plt.bar(race_labels, race_counts, color='skyblue')
    
    # Add percentage labels above each bar
    for bar, percentage in zip(bars, race_percentages):
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width() / 2, height, f'{percentage:.1f}%', ha='center', va='bottom')

    # Add labels and title
    plt.xlabel('Race Categories')
    plt.ylabel('Number of Subjects')
    plt.title('Race Distribution')
    
    # Rotate the x-axis labels for better readability
    plt.xticks(rotation=45, ha="right")
    
    # Display the chart
    plt.tight_layout()
    plt.show()
```

```python
def print_family_id_statistics(df):
    # Group by 'rel_family_id' and filter groups with more than one subject
    grouped = df.groupby('rel_family_id').filter(lambda x: len(x) > 1)
    
    # Count the number of subjects sharing family IDs
    total_shared_subjects = grouped['src_subject_id'].nunique()
    
    # Group again to count the size of each family
    family_sizes = grouped.groupby('rel_family_id').size()
    
    # Count doubles (families with 2 members) and triplets (families with 3 members)
    doubles = (family_sizes == 2).sum()
    triplets = (family_sizes == 3).sum()
    quad = (family_sizes == 4 ).sum()
    cint = (family_sizes == 5 ).sum()
    
    # Print the total number of shared subject IDs
    print(f"Total number of subjects sharing family IDs: {total_shared_subjects}")
    
    # Print the number of doubles and triplets
    print(f"Number of doubles (2 subjects): {doubles}")
    print(f"Number of triplets (3 subjects): {triplets}")
    print(f"Number of  (4 subjects): {quad}")
    print(f"Number of  (5 subjects): {cint}")

    # Optionally, print out the actual groupings
    for family_id, group in grouped.groupby('rel_family_id'):
        print(f"\nFamily ID: {family_id}")
        print("Subject IDs:")
        print(group['src_subject_id'].tolist())
        
print_family_id_statistics(family_df)
```
