```python
import numpy as np 
import pandas as pd
```

```python
qcData = pd.read_csv('/data/neuromark2/Data/ABCD/Data_info/Demo50/abcd-data-release-5.0/core/imaging/mri_y_qc_incl.csv')
cognitionScores = pd.read_csv('/data/neuromark2/Data/ABCD/Data_info/Demo50/abcd-data-release-5.0/core/imaging/mri_y_tfmr_nback_beh.csv')
df_smri = pd.read_csv('/data/users4/sdeshpande8/ROI_Mean_Analysis/smri_roi_values.csv')

attention1 = pd.read_csv('/data/users4/sdeshpande8/Attention_Data/CBCL_Inattention.csv')
attention2 = pd.read_csv('/data/users4/sdeshpande8/Attention_Data/KSAD_Inattention.csv')
attention3 = pd.read_csv('/data/users4/sdeshpande8/Attention_Data/PCA1_Inattention.csv')
```

```python
df_smri
```

```python
new_names = (
    ['F' + str(i+1) for i in range(167)] +  # First 169 features
    ['T' + str(i+1) for i in range(12)] +   # Next 12 features
    ['C' + str(i+1) for i in range(31)] +   # Next 31 features
    ['R1'] +                                # Next 1 feature
    ['P' + str(i+1) for i in range(124)]    # Remaining features
)
```

## QC Filtering

```python
qcData.head()
```

```python
qcData_filtered = qcData[qcData['imgincl_t1w_include']==1]
```

```python
qcData_filtered_baseline = qcData_filtered[qcData_filtered['eventname']=='baseline_year_1_arm_1']
```

```python
qcData_filtered_baseline
```

```python
df_smri.head()
```

```python
df_smri_filtered = df_smri[df_smri['src_subject_id'].isin(qcData_filtered_baseline['src_subject_id'])]
```

```python
df_smri_filtered
```

## Cognition Scores Data

```python
## Consider all baseline_year1 cognitive scores only

cognitionScores_baseline = cognitionScores[cognitionScores['eventname']=='baseline_year_1_arm_1']
print(cognitionScores_baseline.shape)
cognitionScores_baseline.head()
```

```python
#  Keep just the following values: 
#  Processing Speed      i.e tfmri_nb_all_beh_c0b_mrt
#  Response Variability  i.e tfmri_nb_all_beh_c2b_stdrt
#  Attention/Vigilance   i.e tfmri_nb_all_beh_c0b_rate
#  Working Memory        i.e tfmri_nb_all_beh_c2b_rate

cognition_score_targets = cognitionScores_baseline[['src_subject_id', 'tfmri_nb_all_beh_c0b_mrt', 'tfmri_nb_all_beh_c2b_stdrt', 'tfmri_nb_all_beh_c0b_rate', 'tfmri_nb_all_beh_c2b_rate']]
print(cognition_score_targets.shape)
cognition_score_targets.head()
```

```python
merged_df_with_targets = pd.merge(df_smri_filtered, cognition_score_targets, on='src_subject_id', how='left')
```

```python
merged_df_with_targets
```

```python
merged_df_with_targets = merged_df_with_targets.dropna()
```

```python
merged_df_with_targets
```

```python
merged_df_with_targets.to_csv('SMRI_Mean_ROI_Cognition.csv', index = False)
```
