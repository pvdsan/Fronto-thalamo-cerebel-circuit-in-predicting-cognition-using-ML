```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import scipy.io as sio
```

```python
dataPathSMRI = '/data/analysis/collaboration/Multi-sites_genetics/ABCD/roi_features/data_smri_roi_bl.mat'
dataPathDTI = '/data/analysis/collaboration/Multi-sites_genetics/ABCD/roi_features/data_dti_roi_bl.mat'

dataSMRI  = sio.loadmat(dataPathSMRI)
dataDTI  = sio.loadmat(dataPathDTI)
```

## Prepare DTI Data

```python
subjectLabels_DTI = dataDTI['imgpath_dti']
subjectLabelsData_DTI = np.array([item[0][0] for item in subjectLabels_DTI])
data_roi_dti = dataDTI['data_dti_roi']

```

```python
roiname = dataDTI['roiname']
```

```python
roiname
```

```python


df = pd.DataFrame(data_roi_dti, columns=[f"feature_{i}" for i in range(341)])
df.insert(0, "src_subject_id", subjectLabelsData_DTI)  # Add subject array as the first column
df['src_subject_id'] = df['src_subject_id'].str.replace(r'^(NDAR)_?(.+)$', r'\1_\2', regex=True)
df.to_csv('dti_roi_values.csv', index=False)
```
